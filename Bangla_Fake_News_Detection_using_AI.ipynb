```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout, GRU
from tensorflow.keras.models import Model
from transformers import TFDistilBertModel, AutoTokenizer
import numpy as np

# Assuming these are defined from your previous code
max_len = 40
num_classes = 2  # Binary classification (Fake/Real)
distil_bert_model = TFDistilBertModel.from_pretrained('distilbert-base-multilingual-cased')

# Custom Keras layer to wrap TFDistilBertModel
class TFDistilBertLayer(tf.keras.layers.Layer):
    def __init__(self, distil_bert_model, **kwargs):
        super(TFDistilBertLayer, self).__init__(**kwargs)
        self.distil_bert_model = distil_bert_model

    def call(self, inputs):
        input_ids, attention_mask = inputs
        # Ensure inputs are tf.Tensor
        input_ids = tf.cast(input_ids, tf.int32)
        attention_mask = tf.cast(attention_mask, tf.int32)
        outputs = self.distil_bert_model(input_ids=input_ids, attention_mask=attention_mask)
        return outputs[0]  # Return the last hidden state

def create_model():
    # Define inputs
    input_shape = Input(shape=(max_len,), dtype='int32', name='input_ids')
    masks = Input(shape=(max_len,), dtype='int32', name='attention_mask')

    # Wrap DistilBERT in a custom layer
    distil_bert_layer = TFDistilBertLayer(distil_bert_model)([input_shape, masks])

    # Add GRU and dense layers
    X = GRU(128, return_sequences=True)(distil_bert_layer)
    X = tf.keras.layers.GlobalMaxPool1D()(X)
    X = Dense(64, activation='tanh')(X)
    X = Dense(32)(X)
    X = Dense(16)(X)
    X = Dense(8)(X)
    X = Dense(4)(X)
    X = Dropout(0.5)(X)
    X = Dense(num_classes, activation='softmax')(X)

    # Create model
    model = Model(inputs=[input_shape, masks], outputs=X)
    return model

```

```python
# Create and summarize model
model = create_model()
model.summary()
```

```python
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
metric = tf.keras.metrics.SparseCategoricalAccuracy('accuracy')
optimizer = tf.keras.optimizers.Adam(learning_rate=2e-5)
model.compile(loss=loss,optimizer=optimizer, metrics=[metric])
```

```python
import gradio as gr
import numpy as np
from transformers import DistilBertTokenizer
from tensorflow.keras.models import load_model

# Load tokenizer and model
distil_bert_tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
# model = load_model('your_model_path.h5')  # Uncomment and load your model if needed

# Prediction function
def classify_news(text):
    test_input = distil_bert_tokenizer.batch_encode_plus(
        [text],
        add_special_tokens=True,
        max_length=40,
        padding='max_length',
        truncation=True,
        return_attention_mask=True,
        return_tensors='np'
    )

    test_input_ids = np.array(test_input['input_ids'])
    test_input_mask = np.array(test_input['attention_mask'])

    prediction = model.predict([test_input_ids, test_input_mask])
    test_pred_label = np.argmax(prediction, axis=1)[0]

    return "🟢 News is Real" if test_pred_label == 1 else "🔴 News is Fake"

# Gradio interface
interface = gr.Interface(
    fn=classify_news,
    inputs=gr.Textbox(label="Enter News Text", lines=5, placeholder="Type or paste news content here..."),
    outputs=gr.Textbox(label="Prediction"),
    title="Bangla Fake News Classifier",
    description="Enter Bangla news to classify it as Real or Fake using DistilBERT."
)

interface.launch()
```
